```python
df['unit_price'].str.replace('$','')
#It will remove the $ sign from the unit_price column.
```

```python
df.head(5)
#It will give us the first 5 rows of the data after removing the $ sign from the unit_price column.
#after removing the $ sign, the data type of the unit_price column is still object.
#So we need to convert the data type of the unit_price column to float. 
```

```python
df['unit_price'] = df['unit_price'].str.replace('$', '', regex=False).astype(float)
#It will remove the $ sign from the unit_price column and convert the data type of the unit_price column to float.
```

```python
df.head(5)
#It will give us the first 5 rows of the data after removing the $ sign from the unit_price column and converting the data type of the unit_price column to float.
```

```python
df.info()
#It will give us the information about the data after removing the $ sign from the unit_price column and converting the data type of the unit_price column to float.
```
